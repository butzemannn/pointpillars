```python
%matplotlib notebook
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import torch
import numpy as np
import sys
#sys.path.append("/home/qhs67/git/bachelorthesis_sven_thaele/code/")
sys.path.insert(0, "../pointpillars")

from compute_pillars import calculate_pillars
```

```python
class PillarsViewer():
    def __init__(self):
        pass

    def view_pillars(self, pil_points, pil_centers):
        count = torch.count_nonzero(pil_points, dim=2).permute(1,0)
        count = torch.max(count, dim=1)[0].tolist()

        fig = plt.figure()
        ax1 = fig.add_subplot(111, projection='3d')

        n = pil_points.shape[0]
        xpos = pil_centers[:,0].tolist()
        ypos = pil_centers[:,1].tolist()
        zpos = np.zeros(n)

        dx = np.ones(n)
        dy = np.ones(n)
        dz = count

        print(xpos, ypos, zpos, dx, dy, dz)
        ax1.bar3d(xpos, ypos, zpos, dx, dy, dz, color='#00ceaa')
        plt.show()
```

```python
file_location_lid = "/home/qhs67/git/bachelorthesis_sven_thaele/code/data/kitti/training/velodyne/000100.bin"

pc_vel = np.fromfile(file_location_lid, dtype=np.float32, count=-1)
pc_vel = pc_vel.reshape([-1,4])
pc_vel = torch.as_tensor(pc_vel).to("cuda:0")
pil_points, pil_centers = calculate_pillars(pc_vel)

pillarsviewer = PillarsViewer()
pillarsviewer.view_pillars(pil_points, pil_centers)
```
